# Seasonality


Seasonality refers to the trends that occur at specific regular intervals.


```python
# Packages
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

## Downloading Historical Data

```python
ticker= yf.Ticker("EURUSD=X")
```

```python
name= ticker.info["shortName"]
name
```

```python
#dataframes
df= ticker.history(period= "10y") #10year dataframe
df2= ticker.history(period="5y") #5 year dataframe
```

```python
#Show data
df
```

## Seasonality Data

### 10 Year Data

```python
# creating a percentage change column
df["10Y Avg. % Change"]= df["Close"].pct_change()*100
df
```

```python
#group by monthly average percentage change
pc10= df.groupby(df.index.month)[["10Y Avg. % Change"]].mean()
pc10
```

### 5 year data

```python
# Create a percentage change column
df2["5Y Avg. % Change"]= df2["Close"].pct_change()*100

#group by monthly average percentage change
pc5= df2.groupby(df2.index.month)[["5Y Avg. % Change"]].mean()

pc5
```

### Combined Seasonality Data

```python
#combining the 2 sets of data
pct= pd.concat([pc5, pc10], axis= 1)
pct
```

```python
# Setting a month name column as the index
pct['Month']=['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

pct=pct.reset_index()
pct=pct.set_index('Month')
del pct['Date']
pct
```

## Visualising Data

```python
ax= pct.plot(kind='bar', width= 0.7, figsize=(11,8))
ax.set_title(f'{name} Seasonality')
ax.set_ylabel('Average Percentage Change');
```

## Conclusion

The seasonality plot above gives an early look into how the asset typically performs month after month, based on 5 and 10 years of historical data.
While history doesn’t always repeat itself, this information can be a useful tool for pointing out important trends.
